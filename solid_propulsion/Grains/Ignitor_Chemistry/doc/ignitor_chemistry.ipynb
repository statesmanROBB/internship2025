# Thermochemical Characterization of KFBP Ignitor

## What is KFBP?

This is a custom ignitor made of scrap KNSB, ferric oxide and black(gun) powder. We noticed that there is not much documentation on ignitors and we decided to fix that. We'll dive deeply into this topic and hopefully it shall be of intellectual merit.
